## 条件分岐(if文)

```python
x = 100
if  x > 50:
    print("大きい")
else:
    print("小さい")
```

## 繰り返し処理(for文)

繰り返し可能オブジェクト(イテラブルオブジェクト)を直接for文で実行

```python
li = [1, 2, 4]
for i in li:
    print(i)
```

```python
range(10)
```

```python
for i in range(10):
    print(i)
```

## 条件分岐と繰り返し処理の組み合わせ

1から20までの偶数を出力する

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```python
out = []
for i in range(1, 21):
    if i % 2 == 0:
        out.append(i)
out
```

## 同じ内容を内包表記で書く

```python
[i for i in range(1, 21) if i % 2 ==0]
```

## 別の解法

```python
list(range(2, 21, 2))
```
